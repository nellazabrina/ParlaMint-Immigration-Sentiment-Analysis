## 1.1 Extract the Speeches & Speakers

````python
import glob

gb_filenames, es_filenames = [], []
for filename in glob.glob("ParlaMint-GB.TEI/*/*.xml"):
  gb_filenames.append(filename)

for filename in glob.glob("ParlaMint-ES.TEI/*/*.xml"):
  es_filenames.append(filename)
````

````python
import re

date_pattern = re.compile(r"(20[12][0-9]\-(?:0[1-9]|1[0-2])\-(?:0[1-9]|[12][0-9]|3[0-1]))")
````

````python
from lxml import etree
import lxml
````

````python
def get_speakers_speeches(filename):
  tree = etree.parse(filename)
  root = tree.getroot()

  speeches_per_speaker = []
  speakers = []
  speeches = []

  date_speech = date_pattern.search(filename).group(1)

  """
  example:
    <u xml:id="ParlaMint-GB_2015-01-05-commons.u28" ana="#regular" who="#OliverColvile">
        <seg xml:id="ParlaMint-GB_2015-01-05-commons.seg29">I wish you a happy new year, Mr Speaker</seg>
        <seg xml:id="ParlaMint-GB_2015-01-05-commons.seg30">lorem ipsum ...</seg>
    </u>
  """
  for utt in root.findall('.//u', namespaces=root.nsmap):
    segs = utt.findall(".//seg",namespaces=root.nsmap)
    try:
      if "who" not in utt.attrib:
        """
        example: (file ParlaMint-GB.TEI/2021/ParlaMint-GB_2021-11-09-lords.xml)
            <u xml:id="ParlaMint-GB_2021-11-09-lords.u72" ana="#regular">
               <seg xml:id="ParlaMint-GB_2021-11-09-lords.seg73">Miles!</seg>
            </u>
        """

        # if there is no speaker, skip this data
        continue
      temp = utt.attrib['who']
      temp = temp.replace('#', '')
      speakers.append(temp)

      speech = ""
      for seg in segs:
        speech += seg.text + " "
      speech = speech.strip()

      speeches.append(speech)
    except Exception as e:
      print("Key Error")
      print(etree.tostring(utt, pretty_print=True))
      raise Exception

  speeches_per_speaker = zip([date_speech] * len(speakers), speakers, speeches)
  return list(speeches_per_speaker)
````

````python
def get_all_get_speakers_speeches(filenames):
  all_datestr, all_speakers, all_speeches = [], [], []
  for filename in filenames:
    res = get_speakers_speeches(filename)
    for date_str, speaker, speech in res:
      all_datestr.append(date_str)
      all_speakers.append(speaker)
      all_speeches.append(speech)

  return all_datestr, all_speakers, all_speeches

all_datestr, all_speakers, all_speeches = get_all_get_speakers_speeches([filename for filename in gb_filenames if "Schema" not in filename])
gb_speech_df = pd.DataFrame({"date": all_datestr, "speaker": all_speakers, "speech": all_speeches})

all_datestr, all_speakers, all_speeches = get_all_get_speakers_speeches([filename for filename in es_filenames if "Schema" not in filename])
es_speech_df = pd.DataFrame({"date": all_datestr, "speaker": all_speakers, "speech": all_speeches})
````

## 1.2 Extract the Person Information

````python
person_gb_filename = "/content/ParlaMint-GB.TEI/ParlaMint-GB-listPerson.xml"
person_es_filename = "/content/ParlaMint-ES.TEI/ParlaMint-ES-listPerson.xml"

def parse_person_info(filename):
  tree = etree.parse(filename)
  root = tree.getroot()
  namespace_xml = "{http://www.w3.org/XML/1998/namespace}"

  results = {
      "person_id": [],
      "forename": [],
      "surname": [],
      "sex": [],
      "affiliation_start_time": [],
      "affiliation_to_time": [],
      "affiliation_id": [],
      "affiliation_role": [],
  }
  for person_tag in root.findall('.//person', namespaces=root.nsmap):
    pers_id = person_tag.attrib[namespace_xml+"id"]
    pers_forename = person_tag.find("persName/forename", namespaces=person_tag.nsmap).text
    pers_surname = person_tag.find("persName/surname", namespaces=person_tag.nsmap).text
    sex_tag = person_tag.find("sex", namespaces=person_tag.nsmap)
    sex = sex_tag.attrib["value"] if sex_tag is not None else None
    for aff_tag in person_tag.findall('.//affiliation', namespaces=person_tag.nsmap):
      # <affiliation from="2015-05-07" ref="#parliament.HC" role="member" to="2017-05-03"/>
      from_time = aff_tag.attrib["from"] if "from" in aff_tag.attrib else None
      org_id = aff_tag.attrib["ref"] if "ref" in aff_tag.attrib else None
      org_id = org_id.replace("#", "") if org_id else None
      role = aff_tag.attrib["role"]
      to_time = aff_tag.attrib["to"] if "to" in aff_tag.attrib else None
      results["person_id"].append(pers_id)
      results["forename"].append(pers_forename)
      results["surname"].append(pers_surname)
      results["sex"].append(sex)
      results["affiliation_start_time"].append(from_time)
      results["affiliation_to_time"].append(to_time)
      results["affiliation_id"].append(org_id)
      results["affiliation_role"].append(role)

  return results

gb_results = parse_person_info(person_gb_filename)
es_results = parse_person_info(person_es_filename)
gb_person_df = pd.DataFrame.from_dict(gb_results)
es_person_df = pd.DataFrame.from_dict(es_results)
````

## 1.3 Extract the Organizations

````python
org_gb_filename = "/content/ParlaMint-GB.TEI/ParlaMint-GB-listOrg.xml"
org_es_filename = "/content/ParlaMint-ES.TEI/ParlaMint-ES-listOrg.xml"

def parse_org_info(filename):
  tree = etree.parse(filename)
  root = tree.getroot()
  namespace_xml = "{http://www.w3.org/XML/1998/namespace}"

  results = {
    "org_id": [],
    "org_role": [],
    "org_fullName": [],
    "org_abbrName": [],
    "orientation_id": [],
  }

  for org_tag in root.findall('.//org', namespaces=root.nsmap):
    org_id = org_tag.attrib[namespace_xml+"id"]
    org_role = org_tag.attrib["role"]
    org_fullName = org_tag.find('orgName[@full="yes"]', namespaces=org_tag.nsmap).text
    abbrName_tag = org_tag.find('orgName[@full="abb"]', namespaces=org_tag.nsmap)
    org_abbrName = abbrName_tag.text if abbrName_tag is not None else None
    state_tag = org_tag.find('state[@type="politicalOrientation"]', namespaces=org_tag.nsmap)
    if state_tag is not None:
      # for example: <org xml:id="GoGB" role="government">
      # has no political orientation
      orientation_tag = state_tag.find("state", namespaces=state_tag.nsmap)
      orientation_id = orientation_tag.attrib["ana"]
      orientation_id = orientation_id.replace("#", "") if orientation_id else None
    else:
      orientation_id = None

    results["org_id"].append(org_id)
    results["org_role"].append(org_role)
    results["org_fullName"].append(org_fullName)
    results["org_abbrName"].append(org_abbrName)
    results["orientation_id"].append(orientation_id)

  return results

gb_results = parse_org_info(org_gb_filename)
es_results = parse_org_info(org_es_filename)
gb_org_df = pd.DataFrame.from_dict(gb_results)
es_org_df = pd.DataFrame.from_dict(es_results)
````

## 1.4 Extract the Orientations

````python
orientation_gb_filename = "/content/ParlaMint-GB.TEI/ParlaMint-taxonomy-politicalOrientation.xml"
orientation_es_filename = "/content/ParlaMint-ES.TEI/ParlaMint-taxonomy-politicalOrientation.xml"

def parse_orientation_info(filename):
  tree = etree.parse(filename)
  root = tree.getroot()
  namespace_xml = "{http://www.w3.org/XML/1998/namespace}"

  results = {
    "orientation_id": [],
    "orientation_term": [],
    "orientation_desc": [],
  }

  for cat_tag in root.findall('.//category', namespaces=root.nsmap):
    orientation_id = cat_tag.attrib[namespace_xml+"id"]
    term = cat_tag.find('catDesc/term', namespaces=cat_tag.nsmap).text
    desc_tag = cat_tag.find('catDesc', namespaces=cat_tag.nsmap)
    desc = ''.join(desc_tag.itertext())
    results["orientation_id"].append(orientation_id)
    results["orientation_term"].append(term)
    results["orientation_desc"].append(desc)

  return results

gb_results = parse_orientation_info(orientation_gb_filename)
es_results = parse_orientation_info(orientation_es_filename)
gb_orientation_df = pd.DataFrame.from_dict(gb_results)
es_orientation_df = pd.DataFrame.from_dict(es_results)
````

````python
from datetime import datetime
import math

def handle_nan(value):
  if isinstance(value, str):
    if value == "nan":
      value = ""
  elif isinstance(value, float):
    if math.isnan(value):
      value = ""

  return value

def is_date_in_range(check_date, start_date, end_date):
  start_date = str(handle_nan(start_date))
  if start_date:
    start_date = datetime.strptime(start_date, '%Y-%m-%d')
  end_date = str(handle_nan(end_date))
  if end_date:
    end_date = datetime.strptime(end_date, '%Y-%m-%d')
  given_date = str(check_date)
  given_date = datetime.strptime(check_date, '%Y-%m-%d')

  # Check if the given_date is within the range
  if start_date and end_date:
    return start_date <= given_date <= end_date
  elif not start_date:
    # unknown start_date (considered invalid/incomplete information)
    return False
  elif not end_date:
    # assume: the person still not left the party, that's why the end_date is None
    return start_date <= given_date
  else:
    return False
````

````python
import re

right_pattern = re.compile(r"right", re.IGNORECASE)
left_pattern = re.compile(r"left", re.IGNORECASE)
center_pattern = re.compile(r"(center|central|centre)", re.IGNORECASE)

def classify_orientation(text):
  labels = []
  if not text:
    labels.append("NEUTRAL")
  else:
    if right_pattern.search(text):
      labels.append("RIGHT")
    if left_pattern.search(text):
      labels.append("LEFT")
  if not labels:
    if center_pattern.search(text):
      labels.append("CENTER")

  return labels
````

````python
from tqdm import tqdm

orientations_list = []
affiliations_list = []
sex_list = []

with tqdm(total=len(gb_speech_df)) as pbar:
  for idx, row in gb_speech_df.iterrows():
    orientations = []
    affiliations = []
    check_date = row["date"]
    selected_org = gb_person_df[gb_person_df["person_id"] == row["speaker"]]
    for sorg_idx, sorg_row in selected_org.iterrows():
      curr_aff = handle_nan(sorg_row["affiliation_id"])
      if not curr_aff:
        continue
      start_date = sorg_row["affiliation_start_time"]
      end_date = sorg_row["affiliation_to_time"]
      if is_date_in_range(check_date, start_date, end_date):
        affiliations.append(curr_aff)
        selected_orient = gb_org_orientation_df[gb_org_orientation_df["org_id"] == sorg_row["affiliation_id"]]
        if len(selected_orient) > 0:
          desc = selected_orient["orientation_desc"].tolist()
          desc = [item for item in desc if item]
          orientations.extend(desc)
    sex = sorg_row["sex"]
    sex_list.append(sex)

    orientations = list(dict.fromkeys(orientations))
    affiliations = list(dict.fromkeys(affiliations))
    orientations_list.append(";".join(orientations) if orientations else "")
    affiliations_list.append(";".join(affiliations) if affiliations else "")

    pbar.update(1)
````

**Using Word2vec Embeddings To Generate Similar Keywords**  

````python
import gensim.downloader as api

# Pre-trained Word2Vec model
model = api.load("word2vec-google-news-300")

# Spacy Model for Lemmatization
nlp_english = spacy.load("en_core_web_sm")
````

Here we will Lemmatize the seed keywords so that we don't have any duplicates

````python
def lemma_keywords(text):
    doc = nlp_english(text)
    tokens = [token.lemma_ for token in doc]
    preprocessed_text = ' '.join(tokens)
    return preprocessed_text
````

````python
# Lemmatize each keyword
lemmatized_keywords = set(lemma_keywords(keyword) for keyword in seed_keywords)
````

Here we will find the top 5 similar words which are closest to the seed keyword in Word2Vec

````python
similar_words = set()

for seed in lemmatized_keywords:

    try:
      top_similar = model.most_similar(seed, topn=5)

      for word, similarity in top_similar:
          similar_words.add(word)
    except:
      print(f"{seed} NOT FOUND")
````

````python
similar_words = [word.replace("_", " ") for word in similar_words]
similar_words
````

**Using Contextual Word Embeddings to Rank the Similar keyword List**  
Here our goal is to rank the set of similar keywords which we generated in the previous step. TWe will rank using a contextual embedding and then choose the keywords whose similarity lies above our threshold

Here we import the Bert Base tokenizer and Model

````python
from transformers import AutoTokenizer, AutoModel
import torch

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
model = AutoModel.from_pretrained("bert-base-uncased")
````

Here we write a function which calculates the vector word embeddings using BERT

````python
# Function to get embeddings
def get_word_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", add_special_tokens=True)
    outputs = model(**inputs)
    embeddings = outputs.last_hidden_state[:, 1, :].detach().numpy()
    return embeddings
````

````python
# Function to compute cosine similarity
def cosine_similarity(a, b):
    a_flat = a.flatten()
    b_flat = b.flatten()
    return np.dot(a_flat, b_flat) / (np.linalg.norm(a_flat) * np.linalg.norm(b_flat))
````

To Rank the Similar Words, we have two stages.
1.  In stage 1, we calculate the cosine similarity of a similar word with every seed keyword. We then sum all these cosine similarities, of that similar word and then divide it by the total number of seed keywords.
2.  In stage 2, we iterate this process for every similar word and after having the rankings of every similar word we will then sort them into descending order

````python
def RankSimilarWords(keywords, candidates):
  keyword_embeddings = []
  similarities = {}

  # Convert keywords into embeddings
  for word in keywords:
    emb = get_word_embedding(word)
    keyword_embeddings.append(emb)

  # Calculate similarity of all
  for word in candidates:
    embedding = get_word_embedding(word)
    sum = 0
    count = 0
    for keyword in keyword_embeddings:
      sum += cosine_similarity(keyword, embedding)
      count +=1
    similarities[word] = sum/count

  # Sort similarities
  sorted_candidates = sorted(similarities.items(), key=lambda item: item[1], reverse=True)

  return sorted_candidates
````

````python
import pandas as pd
from flair.models import TextClassifier
from flair.data import Sentence
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import confusion_matrix
````

````python
# Initialize the Flair text classifier
classifier = TextClassifier.load('en-sentiment')
````

````python
def get_sentiment_score(text):
  # Truncate the text to fit the maximum input size supported by the model
  max_input_length = 512
  text = text[:max_input_length]

  sentence = Sentence(text)
  classifier.predict(sentence)

  sentiment_label = sentence.labels[0].value
  if sentiment_label == 'POSITIVE':
    sentiment_score = sentence.labels[0].score
  elif sentiment_label == 'NEGATIVE':
    sentiment_score = -sentence.labels[0].score
  else:
    sentiment_score = 0

  # Map sentiment score to a scale from 1 to 5
  if sentiment_score == 0:
    return 3
  else:
    mapped_score = round((sentiment_score + 1) * 2)
    return min(max(mapped_score, 1), 5)
````

### 7.1.1. Data Augmentation

````python
# Define the augmenter
augmenter = ContextualWordEmbsAug(model_path='bert-base-uncased', action="substitute")
````

````python
numbers = dat_20["label"].value_counts().to_dict()
numbers.values()
````

````python
list(numbers.values())
````

````python
numbers = dat_20["label"].value_counts().to_dict()
# Calculate the LCM/KPK
lcm_result = np.lcm.reduce(list(numbers.values()))

all_text, all_labels = [], []
for _, row in dat_20.iterrows():
  text = row["text"]
  label = row["label"]
  all_text.append(text)
  all_labels.append(label)

  # Augment the text multiple times
  num_augmentations = int(lcm_result/numbers[label]) - 1
  for _ in range(num_augmentations):
      augmented_text = augmenter.augment(text)
      augmented_text = augmented_text[0] if isinstance(augmented_text, list) else augmented_text
      all_text.append(augmented_text)
      all_labels.append(label)

dat_20_aug = pd.DataFrame()
dat_20_aug["text"] = all_text
dat_20_aug["label"] = all_labels
````

````python
numbers = dat_10["label"].value_counts().to_dict()
# Calculate the LCM/KPK
lcm_result = np.lcm.reduce(list(numbers.values()))

all_text, all_labels = [], []
for _, row in dat_10.iterrows():
  text = row["text"]
  label = row["label"]
  all_text.append(text)
  all_labels.append(label)

  # Augment the text multiple times
  num_augmentations = int(lcm_result/numbers[label]) - 1
  for _ in range(num_augmentations):
      augmented_text = augmenter.augment(text)
      augmented_text = augmented_text[0] if isinstance(augmented_text, list) else augmented_text
      all_text.append(augmented_text)
      all_labels.append(label)

dat_10_aug = pd.DataFrame()
dat_10_aug["text"] = all_text
dat_10_aug["label"] = all_labels
````

## 7.3. Polarity Each Word

````python
def get_all_sentiment_polarities(sentence):
  # Tokenize the sentence into words
  words = sentence.split()

  # Calculate sentiment polarity for each word
  word_polarities = {}
  for word in words:
      # Get synsets for the word
      synsets = wn.synsets(word)
      # Initialize positive and negative scores
      pos_score = 0
      neg_score = 0
      # Accumulate sentiment scores from all synsets
      for synset in synsets:
          synset_pos_score = swn.senti_synset(synset.name()).pos_score()
          synset_neg_score = swn.senti_synset(synset.name()).neg_score()
          pos_score += synset_pos_score
          neg_score += synset_neg_score
      # Calculate average sentiment polarity for the word
      if len(synsets) > 0:
          avg_pos_score = pos_score / len(synsets)
          avg_neg_score = neg_score / len(synsets)
          polarity = avg_pos_score - avg_neg_score
          word_polarities[word] = polarity

  # Print sentiment polarity for each word
  # for word, polarity in word_polarities.items():
  #     print(f"{word}: {polarity}")

  return word_polarities
````

````python
result = {
    "text": [],
    "polarities": [],
    "mean": [],
    "std": [],
    "max": [],
    "min": [],
}
for sentence in kw_sentences:
  result["text"].append(sentence)
  polarities = get_all_sentiment_polarities(sentence)
  result["polarities"].append(polarities)
  numbers = [pol for pol in polarities.values()]
  result["mean"].append(np.mean(numbers))
  result["std"].append(np.std(numbers))
  result["max"].append(np.max(numbers))
  result["min"].append(np.min(numbers))
````
